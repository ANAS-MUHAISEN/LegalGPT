```python
import pdfplumber

def extract_text_from_pdf(file_path):
    full_text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            full_text += page.extract_text() + "\n"
    return full_text
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_contract(text):
    summary = summarizer(text, max_length=300, min_length=100, do_sample=False)
    return summary[0]['summary_text']
```

```python
import streamlit as st

st.title("📄 المحامي الذكي – تحليل وتلخيص العقد")

uploaded_file = st.file_uploader("ارفع العقد (PDF)", type="pdf")

if uploaded_file:
    with open("temp.pdf", "wb") as f:
        f.write(uploaded_file.read())

    contract_text = extract_text_from_pdf("temp.pdf")
    st.subheader("📃 النص المستخرج:")
    st.write(contract_text[:1000] + "...")

    if st.button("🔍 لخص العقد"):
        summary = summarize_contract(contract_text)
        st.subheader("📌 ملخص العقد:")
        st.write(summary)
```

```python
import streamlit as st
import os
from pdfplumber import open as pdf_open
import re

def extract_text_from_pdf(file_path):
    text = ""
    with pdf_open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text.strip()

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[®™]', '', text)
    return text.strip()

st.set_page_config(page_title="المحامي الذكي", layout="centered")
st.title("📚 المحامي الذكي - مساعدك القانوني")

uploaded_file = st.file_uploader("📄 حمّل مستند قانوني بصيغة PDF", type=["pdf"])

if uploaded_file:
    file_path = os.path.join("uploads", uploaded_file.name)
    os.makedirs("uploads", exist_ok=True)
    with open(file_path, "wb") as f:
        f.write(uploaded_file.read())

    st.success("✅ تم تحميل الملف بنجاح")

    if st.button("🔍 استخراج النص"):
        raw_text = extract_text_from_pdf(file_path)
        cleaned = clean_text(raw_text)
        st.text_area("📃 النص القانوني المستخرج والمنظف:", value=cleaned, height=400)
```

```python
import streamlit as st
from transformers import pipeline
import os
import pdfplumber
import re

def extract_text_from_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text.strip()

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[®™]', '', text)
    return text.strip()

@st.cache_resource
def load_qa_model():
    return pipeline("question-answering", model="deepset/roberta-base-squad2")

qa_pipeline = load_qa_model()

st.title("📚 المحامي الذكي - مساعدك القانوني")

uploaded_file = st.file_uploader("📄 حمّل مستند قانوني PDF", type=["pdf"])

if uploaded_file:
    os.makedirs("uploads", exist_ok=True)
    file_path = os.path.join("uploads", uploaded_file.name)
    with open(file_path, "wb") as f:
        f.write(uploaded_file.read())
    st.success("✅ تم تحميل الملف بنجاح")

    if st.button("🔍 استخراج النص"):
        raw_text = extract_text_from_pdf(file_path)
        cleaned = clean_text(raw_text)
        st.text_area("📃 النص القانوني المستخرج والمنظف:", value=cleaned, height=400)

        question = st.text_input("❓ اسأل المحامي الذكي سؤالاً عن المستند:")

        if question:
            result = qa_pipeline(question=question, context=cleaned)
            st.markdown(f"**الإجابة:** {result['answer']}")
```
